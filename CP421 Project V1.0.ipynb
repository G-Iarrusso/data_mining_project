```python
import numpy
import pandas as pd
```

```python
data = pd.read_csv('Testing.csv')
```

```python
pos = ['OT','OL','OG','C']
training_data = data[data['Pos'].isin(pos)]
```

```python
training_data = training_data.drop(['Player', 'School','College','Ht'], axis=1)

cond = training_data['Drafted'].isnull()
training_data.loc[cond,'Drafted'] = 'Not Drafted'
cond = training_data['Drafted'] != 'Not Drafted'
training_data.loc[cond,'Drafted'] = 'Drafted'
training_data.describe()
```
